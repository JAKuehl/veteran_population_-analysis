This will change values to strings so that they will displayy better in Tableau.

```python
# Import libraries
import pandas as pd
import scipy as sc
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import time

# Read in Data
vet_1 = pd.read_csv("./tableau_data/vets_1yr_tableau.csv")
pop_1 = pd.read_csv("./tableau_data/us_pop_1yr_tableau.csv")
# vet_5 = pd.read_csv("../working_data/vets_5yr_tableau.csv")
# pop_5 = vet_1 = pd.read_csv("../working_data/us_pop_5yr_tableau.csv")

'''Additional viewing options'''
# This code sets the notebook to display maximum columns.  Uncomment it to see trunacted view
pd.set_option('display.max_columns', len(pop_1))  

# This resets the view to a truncated output
#pd.reset_option('display.max_rows')   
```

```python
# To maximize Tableau certain features need to be remapped to strings.  This means remapping almost the entire dataset.
def title_data(df):
    
    # Convert nulls back
    df = df.replace({-99:np.NaN})
    
    # Building dictionaries for the various values
    division = {0:'Puerto Rico', 1:'New England', 2:'Middle Atlantic', 3:'East North Central',
                4:'West North Central', 5:'South Atlantic', 6:'East South Central', 
                7:'West South Central', 8:'Mountain', 9:'Pacific'}
    region = {1:'Northeast', 2:'Midwest', 3:'South', 4:'West', 9:'Puerto Rico'}
    state = {2:'Alaska', 1:'Alabama', 5:'Arkansas', 4:'Arizona', 6:'California', 8:'Colorado',
             9:'Connecticut', 11:'District of Columbia', 10:'Delaware', 12:'Florida', 13:'Georgia',
             15:'Hawaii', 19:'Iowa', 16:'Idaho', 17:'Illinois', 18: 'Indiana', 20:'Kansas', 21:'Kentucky',
             22:'Louisiana', 25: 'Massachusetts', 24:'Maryland', 23: 'Maine', 26: 'Michigan', 27: 'Minnesota',
             29:'Missouri', 28:'Mississippi', 30:'Montana', 37: 'North Carolina', 38: 'North Dakota', 31:'Nebraska',
             33:'New Hampshire', 34:'New Jersey', 35:'New Mexico', 32:'Nevada', 36:'New York', 39: 'Ohio',
             40:'Oklahoma', 41:'Oregon', 42:'Pennsylvania', 72:'Puerto Rico', 44:'Rhode Island',
             45:'South Carolina', 46:'South Dakota', 47:'Tennessee', 48:'Texas', 49: 'Utah', 51:'Virginia',
             50:'Vermont', 53:'Washington', 55:'Wisconsin', 54:'West Virginia', 56:'Wyoming'}
    citizen = {1:'Born in US', 2:'Born in PR, MP, VI', 3:'Born Abroad, US parents', 4:'Naturlaized', 5:'Not Citizen'}
    worker = {np.NaN: 'Never Worked', 1:'Private, for-profit', 2:'Private, non-profit', 3:'Local Govt', 4:'State Govt', 
              5:'Fed Govt', 6:'Self Employ, not inc', 7:'Self Employ, inc', 8:'Work, no pay', 9:'Unemployed'}
    yes0 = {0:'No', 1:'Yes'}
    yes1 = {1:'Yes', 2:'No'}
    yes2 = {1:'No', 1:'Yes'}
    disablity = {np.NaN:'Not Reported', 1:'0%', 2:'10-20%', 3:'30-40%', 4:'50-60%', 5:'70-100%', 6:'0 or Never Served'}
    english = {1:'Fluent', 2:'Well', 3:'Not Well', 4:'None'}
    car = {0:'No Car', 1:'Drives Alone', 2:'Carpool'}
    mar = {1:'Married', 2:'Widowed', 3:'Divorced', 4:'Seperated', 5:'Never Married'}
    yes_mar = {0:'Never Married', 1:'Yes', 2:'No'}
    num_mar = {0:'Never Married', 1:'1 Time', 2:'2 Times', 3:'3 or more'}
    military = {1:'Active Duty', 2:'Active Duty(past)', 3:'Reserve/Natl Guard', 0:'Never served'}
    era = {np.NaN:'No Active Duty', 0:'Did NOT Serve Period', 1:'Served Period'}
    wk_look = {0:'At Work', 1:'Looking for Work', 2:'NOT Looking for Work'}
    school = {1:'No Attendence in last 3', 2:'Attending Public', 3:'Attending Private'}
    education = {1:'None', 2:'Preschool', 3:'Kindergarten', 4:'Grade 1', 5:'Grade 2', 6:'Grade 3', 7:'Grade 4', 8:'Grade 5', 
                 9:'Grade 6', 10:'Grade 7', 11:'Grade 8', 12:'Grade 9', 13:'Grade 10', 14:'Grade 11', 
                 15:'Grade 12, no diploma', 16:'Diploma', 17:'GED', 18:'Some College, <1 year', 19:'Some College,no degree',
                 20:'Associates', 21:'Bachelors', 22:'Masters', 23:'Professional Degree', 24:'Doctorate'}
    sex = {1:'Male', 2:'Female'}
    hours = {0:'Not Working', 1:'1-98 hours', 2:'99+ hours'}
    work = {1:'In Last Year', 2:'1-5 Years Ago', 0:'5+ Years, Never Worked'}
    weeks_wk = {np.NaN:'Not working', 1:'50-52 Weeks', 2:'48-49 Weeks', 3:'40-47 Weeks', 
                4:'27-39 Weeks', 5:'14-26 Weeks', 6:'>14 Weeks'}
    employ = {1:'Civilian, At Work', 2:'Civilian, NOT At Work', 3:'Unemployed', 4:'Armed Forces, At Work', 
              5:'Armed Forces, NOTAt Work', 7:'Not in Labor Force'}
    disability2 ={1:'With Disability', 2:'Without Disability'}
    coverage = {1:'Has Health Insurance', 2:'No Health Insurance'}
    hisp = {0:'Not Hispanic', 1:'Hispanic'}
    mar2 = {1:'Married, spouse present', 2:'Married, spouse absent', 3:'Widowed', 4:'Divorced', 5:'Seperated', 6:'Never Married'}
    native = {1:'Native', 2:'Foreign Born'}
    race = {1:'White', 2:'Black', 3:'Native Tribe', 4:'Alaskan Tribe', 5:'General Native Tribe', 6:'Asian', 7:'Hawiian or PI',
           8:'Other Race', 9:'Mixed Race'}
    service = {np.NaN:'No Active Duty', 1:'Gulf War: 9/2001 or later', 2:'Gulf War: 9/2001 or later and Gulf War: 8/1990 - 8/2001', 
               3:'Gulf War: 9/2001 or later and Gulf War: 8/1990 - 8/2001 and Vietnam Era', 4:'Gulf War: 8/1990 - 8/2001',
               5:'Gulf War: 8/1990 - 8/2001 and Vietnam Era', 6:'Vietnam Era', 7:'Vietnam Era and Korean War', 
               8:'Vietnam Era, Korean War, and WWII', 9:'Korean War', 10:'Korean War and WWII', 
               11:'WWII', 12:'Between Gulf War and Vietnam Era only', 13:'Between Vietnam Era and Korean War only',
               14:'Between Korean War and World War II only', 15:'Pre-WWII only'}
    
    
    # Now for the remapping of all of the features
    df['DIVISION'] = vet_1['DIVISION'].map(division).fillna(df['DIVISION'])
    df['REGION'] = vet_1['REGION'].map(region).fillna(df['REGION'])
    df['ST'] = vet_1['ST'].map(state).fillna(df['ST'])
    df['CIT'] = vet_1['CIT'].map(citizen).fillna(df['CIT'])
    df['COW'] = vet_1['COW'].map(worker).fillna(df['COW'])
    df['DDRS'] = vet_1['DDRS'].map(yes1).fillna(df['DDRS'])
    df['DEAR'] = vet_1['DEAR'].map(yes1).fillna(df['DEAR'])
    df['DEYE'] = vet_1['DEYE'].map(yes1).fillna(df['DEYE'])
    df['DOUT'] = vet_1['DOUT'].map(yes1).fillna(df['DOUT'])
    df['DPHY'] = vet_1['DPHY'].map(yes1).fillna(df['DPHY'])
    df['DRAT'] = vet_1['DRAT'].map(disablity).fillna(df['DRAT'])
    df['DRATX'] = vet_1['DRATX'].map(yes1).fillna(df['DRATX'])
    df['DREM'] = vet_1['DREM'].map(yes1).fillna(df['DREM'])
    df['ENG'] = vet_1['ENG'].map(english).fillna(df['ENG'])
    df['HINS1'] = vet_1['HINS1'].map(yes1).fillna(df['HINS1'])
    df['HINS2'] = vet_1['HINS2'].map(yes1).fillna(df['HINS2'])
    df['HINS3'] = vet_1['HINS3'].map(yes1).fillna(df['HINS3'])
    df['HINS4'] = vet_1['HINS4'].map(yes1).fillna(df['HINS4'])
    df['HINS5'] = vet_1['HINS5'].map(yes1).fillna(df['HINS5'])
    df['HINS6'] = vet_1['HINS6'].map(yes1).fillna(df['HINS6'])
    df['JWRIP'] = vet_1['JWRIP'].map(car).fillna(df['JWRIP'])
    df['LANX'] = vet_1['LANX'].map(yes1).fillna(df['LANX'])
    df['MAR'] = vet_1['MAR'].map(mar).fillna(df['MAR'])
    df['MARHT'] = vet_1['MARHT'].map(num_mar).fillna(df['MARHT'])
    df['MARHW'] = vet_1['MARHW'].map(yes_mar).fillna(df['MARHW'])
    df['MARHD'] = vet_1['MARHD'].map(yes_mar).fillna(df['MARHD'])
    df['MIL'] = vet_1['MIL'].map(military).fillna(df['MIL'])
    df['MLPA'] = vet_1['MLPA'].map(era).fillna(df['MLPA'])
    df['MLPB'] = vet_1['MLPB'].map(era).fillna(df['MLPB'])
    df['MLPCD'] = vet_1['MLPCD'].map(era).fillna(df['MLPCD'])
    df['MLPE'] = vet_1['MLPE'].map(era).fillna(df['MLPE'])
    df['MLPFG'] = vet_1['MLPFG'].map(era).fillna(df['MLPFG'])
    df['MLPH'] = vet_1['MLPH'].map(era).fillna(df['MLPH'])
    df['MLPI'] = vet_1['MLPI'].map(era).fillna(df['MLPI'])
    df['MLPJ'] = vet_1['MLPJ'].map(era).fillna(df['MLPJ'])
    df['MLPK'] = vet_1['MLPK'].map(era).fillna(df['MLPK'])
    df['NWLK'] = vet_1['NWLK'].map(wk_look).fillna(df['NWLK'])
    df['SCH'] = vet_1['SCH'].map(school).fillna(df['SCH'])
    df['SCHL'] = vet_1['SCHL'].map(education).fillna(df['SCHL'])
    df['SEX'] = vet_1['SEX'].map(sex).fillna(df['SEX'])
    df['WKHP'] = vet_1['WKHP'].map(hours).fillna(df['WKHP'])
    df['WKL'] = vet_1['WKL'].map(work).fillna(df['WKL'])
    df['WRK'] = vet_1['WRK'].map(yes1).fillna(df['WRK'])
    df['DIS'] = vet_1['DIS'].map(disability2).fillna(df['DIS'])
    df['ESR'] = vet_1['ESR'].map(employ).fillna(df['ESR'])
    df['HICOV'] = vet_1['HICOV'].map(coverage).fillna(df['HICOV'])
    df['HISP'] = vet_1['HISP'].map(hisp).fillna(df['HISP'])
    df['MSP'] = vet_1['MSP'].map(mar2).fillna(df['MSP'])
    df['NATIVITY'] = vet_1['NATIVITY'].map(native).fillna(df['NATIVITY'])
    df['OC'] = vet_1['OC'].map(yes0).fillna(df['OC'])
    df['RAC1P'] = vet_1['RAC1P'].map(race).fillna(df['RAC1P'])
    df['RACAIAN'] = vet_1['RACAIAN'].map(yes0).fillna(df['RACAIAN'])
    df['RACBLK'] = vet_1['RACBLK'].map(yes0).fillna(df['RACBLK'])
    df['RACASN'] = vet_1['RACASN'].map(yes0).fillna(df['RACASN'])
    df['RACNUM'] = vet_1['RACNUM'].map(yes0).fillna(df['RACNUM'])
    df['RACSOR'] = vet_1['RACSOR'].map(yes0).fillna(df['RACSOR'])
    df['RACWHT'] = vet_1['RACWHT'].map(yes0).fillna(df['RACWHT'])
    df['SCIENGP'] = vet_1['SCIENGP'].map(yes1).fillna(df['SCIENGP'])
    df['SCIENGRLP'] = vet_1['SCIENGRLP'].map(yes1).fillna(df['SCIENGRLP'])
    df['VPS'] = vet_1['VPS'].map(service).fillna(df['VPS'])
    df['FCITP'] = vet_1['FCITP'].map(yes0).fillna(df['FCITP'])
    df['FHINS5P'] = vet_1['FHINS5P'].map(yes0).fillna(df['FHINS5P'])
    df['FHINS6P'] = vet_1['FHINS6P'].map(yes0).fillna(df['FHINS6P'])
    df['FMARHYP'] = vet_1['FMARHYP'].map(yes0).fillna(df['FMARHYP'])
    df['FMILPP'] = vet_1['FMILPP'].map(yes0).fillna(df['FMILPP'])
    df['FMILSP'] = vet_1['FMILSP'].map(yes0).fillna(df['FMILSP'])
    df['FPUBCOVP'] = vet_1['FPUBCOVP'].map(yes0).fillna(df['FPUBCOVP'])
    df['WKW'] = vet_1['WKW'].map(weeks_wk).fillna(df['WKW'])
    
    return df
```
